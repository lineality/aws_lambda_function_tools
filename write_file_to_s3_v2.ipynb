```python
"""
write_file_to_s3

This is a function that 'writes' a file to AWS-s3.

The use of try-except is recommended, to handle exceptions etc.
"""

import json
import os
import boto3
from datetime import datetime

def lambda_handler(event, context):

    #############################################
    # Setup, inputs, connections, variables, etc.
    #############################################

    # Get POST items from hitting endpoint
    ITEM_ONE = event['ITEM_ONE'] 
    ITEM_TWO = event['ITEM_TWO'] 

    # Timestamp
    timestamp = datetime.now()

    # Pick your AWS S3 bucket
    s3_bucket_name = 'NAME_OF_S3_BUCKET'

    # file_destination_in_s3
    file_destination_in_s3 = "test_folder/"


    ###################
    # Connecting to S3
    ###################

    # Make "resource"
    s3 = boto3.resource('s3')
    
    # Set your S3 bucket
    bucket = s3.Bucket( s3_bucket_name )
    

    ####################
    # Content to write
    ####################

    # start with a comma, end with an escape character new line
    write_this_text = f",{ITEM_ONE},{ITEM_TWO},{timestamp}\n"


    #################
    # File to write
    #################

    new_file_name = 'NAME_OF_YOUR_FILE.txt'
    aws_location_file_name = f'/tmp/{new_file_name}'
    s3_file_destination_and_file = file_destination_in_s3 + new_file_name

    # safe file in /tmp/ directory because AWS requires 
    with open( aws_location_file_name, 'w') as data:
        data.write( write_this_text )


    #######################################
    # Upload: Write file to S3 your bucket
    #######################################

    bucket.upload_file(aws_location_file_name, s3_file_destination_and_file)
    
    # remove file from temp to prevent issues because AWS is buggy buggy :(
    os.remove( aws_location_file_name ) 


    #########
    # Finish
    #########

    return {
        'statusCode': 200,
        'body': json.dumps('File Created, OK!')
    }
```
